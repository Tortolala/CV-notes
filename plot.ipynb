```python
import numpy
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img_color = cv.imread('./lenna.jpg', cv.IMREAD_COLOR)
img_gray = cv.imread('./lenna.jpg', cv.IMREAD_GRAYSCALE)
```

```python
def imgview(img, title = None, filename = None):
    """ Description of this function.

    Args:
        img (numpy array): image array
        title (string): image title
        filename (string): image filename

    Returns:
        result (type): desc
    """

    if len(img.shape) > 2: 
        plt.imshow(img_color)
    else:
        plt.imshow(img_gray,cmap='gray', vmin = 0, vmax = 255)
        
     
    if (filename != None):
    plt.imsave(fname = filename + ".png", arr = img_gray)
```

```python
imgview(img_gray, title = "a", filename = "aska")
```
