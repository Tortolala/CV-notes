```python
import numpy
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def split_rgb(img, filename = None): 
    fig = plt.figure(figsize=(10, 10))
    ax1 = fig.add_subplot(221)
    ax1.imshow(img)
    ax1.set_title('RGB')
    plt.axis('off')
    ax2 = fig.add_subplot(222)
    ax2.imshow(img[:,:,0], cmap='gray', vmin=0, vmax=255)
    ax2.set_title('R')
    plt.axis('off')
    ax3 = fig.add_subplot(223)
    ax3.imshow(img[:,:,1], cmap='gray', vmin=0, vmax=255)
    ax3.set_title('G')
    plt.axis('off')
    ax4 = fig.add_subplot(224)
    ax4.imshow(img[:,:,2], cmap='gray', vmin=0, vmax=255)
    ax4.set_title('B')
    plt.axis('off')
    if (filename != None):
        plt.savefig(filename + ".png")
    plt.show()
```

```python
img_color = cv.imread('./lenna.jpg', cv.IMREAD_COLOR)
img = cv.cvtColor(img_color, cv.COLOR_BGR2RGB)
```

```python
split_rgb(img, "pruebaaa!")
```

```python
def hist(img, filename = None):
    
    fig = plt.figure(figsize=(10, 10))
    ax1 = fig.add_subplot(111)
    colors = ['r', 'g', 'b']
    for i, color in enumerate(colors):
        histr = cv.calcHist([img], [i], None, [255], [0, 255])
        ax1.plot(histr, c = color)
    plt.show()
```

```python
hist(img)
```
